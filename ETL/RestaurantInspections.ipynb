```python
import pandas as pd
from sqlalchemy import create_engine
```

Extract CSVs into DataFrames

```python
CHIInspection = "C:/Users/bvkka/Desktop/ETLProject/CHIInspection.csv"
CHIInspection_df = pd.read_csv(CHIInspection)
CHIInspection_df.head()
```

```python
SFInspection = "C:/Users/bvkka/Desktop/ETLProject/SFInspections.csv"
SFInspection_df = pd.read_csv(SFInspection)
SFInspection_df.head()
```

 Transform CHI DataFrame

```python
# Create a filtered dataframe from specific columns
CHI_cols = ["Inspection ID","DBA Name","Inspection Date","Inspection Type","Violations","Risk"]
CHI_transformed= CHIInspection_df[CHI_cols].copy()

# Rename the column headers
CHI_transformed = CHI_transformed.rename(columns={        "Inspection ID":"ID",
                                                          "DBA Name": "Business Name",
                                                          "Facility Type": "Facility",
                                                           "Violations": "Violation",
                                                          })

# Clean the data by dropping duplicates and setting the index
CHI_transformed.drop_duplicates("Business Name", inplace=True)
CHI_transformed.set_index("ID", inplace=True)

CHI_transformed.head()
```

 Transform SF DataFrame

```python
# Create a filtered dataframe from specific columns
SF_cols = ["inspection_id","business_name","inspection_date","inspection_type","violation_description","risk_category"]
SF_transformed= SFInspection_df[SF_cols].copy()


# Rename the column headers
SF_transformed = SF_transformed.rename(columns={        "inspection_id":"ID",
                                                          "business_name": "Business Name",
                                                          "inspection_date": "Inspection Date",
                                                          "inspection_type":"Inspection Type",
                                                          "violation_description": "Violation",
                                                          "risk_category":"Risk"
                                                          })


# Clean the data by dropping duplicates and setting the index
SF_transformed.drop_duplicates("Business Name", inplace=True)
SF_transformed.set_index("ID", inplace=True)

SF_transformed.head()
```
